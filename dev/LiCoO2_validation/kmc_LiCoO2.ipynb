# Test on $LiCoO_2$

First We will have to add a dummy atom in the center of 3-Li-triangle, as the center of cluster expansion

```python
from kmcpy.external.pymatgen_structure import Structure
LiCoO2=Structure.from_cif("EntryWithCollCode51182.cif")
LiCoO2.make_supercell([2,2,1])
import numpy as np
#np.set_printoptions(precision=2)
for site in LiCoO2:
    if "Li+" in site:
        print(site,site.properties,site.frac_coords)
print(LiCoO2)
```

Here the 1st, 3rd, 4th Li+ obviously form a triangle, refer to Anton's paper (Van Der Ven A, Ceder G. Lithium migration in layered LixCoO2 Electrochemical and Solid-State Letters. 3: 301-304. DOI: 10.1149/1.1391130), the center of local cluster expansion is the center of such Lithium triangle. Then we need to add an dummy atom in the center for events generator

```python
from kmcpy.external.pymatgen_structure import Structure
import numpy

def add_dummy(structure1,structure2,site1,site2,site3):
    structure1.append(species="Na+",coords=(structure2[site1].coords+structure2[site2].coords+structure2[site3].coords)/3,coords_are_cartesian=True,properties={'local_index':0, 'label': 'dummy'})

LiCoO2_2=Structure.from_cif("EntryWithCollCode51182.cif")

for center_Li in [[0,2,3],[0,1,3],[4,6,7],[4,5,7],[8,10,11],[8,9,11]]:
    add_dummy(LiCoO2_2,LiCoO2,*center_Li)

print(LiCoO2_2)
LiCoO2_2.to("cif","LiCoO2_dummy_added.cif",symprec=0.01)
```

```python
mobile_ion_identifier_type="label"
mobile_ion_specie_1_identifier="Na0"
mobile_ion_specie_2_identifier="Li1"
prim_cif_name="LiCoO2_dummy_added.cif"
local_env_cutoff_dict={('Na+','Li+'):4.31}
from kmcpy.event_generator import generate_events3
generate_events3(prim_cif_name=prim_cif_name,local_env_cutoff_dict=local_env_cutoff_dict,mobile_ion_identifier_type=mobile_ion_identifier_type,mobile_ion_specie_1_identifier=mobile_ion_specie_1_identifier,mobile_ion_specie_2_identifier=mobile_ion_specie_2_identifier,species_to_be_removed=["O2-","O","Co+","Co"],distance_matrix_rtol=0.01,distance_matrix_atol=0.01,find_nearest_if_fail=False,convert_to_primitive_cell=False,export_local_env_structure=True,supercell_shape=[2,2,2],event_fname="events.json",event_kernal_fname='event_kernal.csv',verbosity="INFO")
```
